```python
import xlrd
import numpy as np

def readExcel(filename):
    data = xlrd.open_workbook(filename)
    table = data.sheets()[0]            # 開啟第一張表
    return table
```

```python
datastream = readExcel('datastream_sample.xlsx')

company = []
for i in range(1,51):
    temp = [datastream.row_values(i)[2], int(datastream.row_values(i)[4])]
    company.append(temp)
# print(company)

file = []
for i in range(50):
    file.append('_' + str(company[i][1]) + '.TW.xlsx')
# print(file[0])

data = []
for i in range(50):
    temp = readExcel(file[i])
    data.append(temp)
# print(data[0])
# print(data[1])
# print(data.row_values(1)[0])
# for i in range(data.nrows):
    # print(time.localtime(data.row_values(i)[0]))
    # for j in range(1, data.nrows + 1):
```

```python
# data[0].row_values(1)[0]
# print(data[0].row_values(1)[1])
# print(data[49].row_values(1)[1])

indicators = []
for i in range(50):
    date = data[i].row_values(1)[0]
    total = count = 0
    indicator = []
    for j in range(1, data[i].nrows):
        
        if(data[i].row_values(j)[0] >= date + 30):
            date = data[i].row_values(j)[0]
            if(count == 0):
                indicator.append(-1)
            else:
                indicator.append(total / count)
            total = count = 0
        # print("i = ", i, " j = ", j, " data = ", data[i].row_values(j)[1])
        
        if(type(data[i].row_values(j)[1]) != type('0')):
            total = total + data[i].row_values(j)[1]
            count = count + 1
    indicators.append(indicator)
    
print(indicators)
```

```python
grow_rates = []
for i in range(50):
    count = total = 0
    grow_rate = []
    for j in range(1, len(indicators[i])):
        if(indicators[i][j] != -1):
            if(indicators[i][j - 1] != -1):
                count = count + 1
                grow_rate.append(indicators[i][j] / indicators[i][j - 1])
                total = total + indicators[i][j] / indicators[i][j - 1]
    if(count == 0):
        grow_rates.append(-1)
    else:
        grow_rates.append(total / count)
print(grow_rates)
```

```python
prices = []
for i in range(50):
    count = total = 0
    price = []
    for j in range(len(indicators[i])):
        if(indicators[i][j] != -1):
                count = count + 1
                price.append(indicators[i][j])
                total = total + indicators[i][j]
    if(count == 0):
        prices.append(-1)
    else:
        prices.append(total / count)
print(prices)
```

```python
top5 = []
for i in range(5):
    top5.append([0, 0, 0])

for i in range(50):
    for j in range(5):
        if(grow_rates[i] > top5[j][2]):
            top5[j] = [company[i][1], company[i][0], grow_rates[i]]
            break
print(top5)
```

```python
top5 = []
for i in range(5):
    top5.append([0, 0, 0, 0])

for i in range(50):
    for j in range(5):
        if(grow_rates[i] < 1):
            break;
        if(grow_rates[i] * prices[i] > top5[j][1] * top5[j][2]):
            top5[j] = [company[i][1], grow_rates[i], prices[i], i]
            break
print(top5)
```
